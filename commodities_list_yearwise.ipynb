```python
import pandas as pd
```

```python
dict={}

for i in range(7):
    s = 'd_'+str(2012+i)
    dict[s] = pd.read_csv(r'../input/cropyears - '+str(2012+i)+'.csv')
```

```python
len(list(dict['d_2018']['2018']))
```

```python
years = ['2018' for i in range(446)]
df = pd.DataFrame()
df['crops'] = list(dict['d_2018']['2018'])
df['years'] = years
df
```

```python

for i in range(6):
    j = str(2018-i-1)
    for crop in list(dict['d_'+j][j]):
        if crop in list(df.crops):
            t = df[df['crops'] == crop].index
            df.iloc[t,1] = j+', '+df.iloc[t,1]
        else:
            new_row = {'crops':crop, 'years': j}
            df = df.append(new_row, ignore_index=True)


df
```

```python
df = df.sort_values(['crops'])
df
```

```python
df.to_csv('crop_years.csv',index=False)
```
