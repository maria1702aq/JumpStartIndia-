```python
import pandas as pd
```

1. Store each year dataset into a dictionary
2. Remove NaN value, sort columns and sort rows in all data sets 
3. Make a another dataframe and remove the marketcentre column 
4. Group and sum that dataframe
5. Merge this dataframe with the orignal dataframe
6. Create additional column = Market Share : in orginal dataframe
7. Append all year dataframes together
8. Make statewise annual and nationwise annual reports




```python
dict1 = {}


for i in range(9):
    dict1[str(2012+i)]=pd.read_csv(r'../input/'+str(2012+i)+'.csv')
    dict1[str(2012+i)] = dict1[str(2012+i)].dropna()
    dict1[str(2012+i)] = dict1[str(2012+i)][['# Commodity', 'State' , 'Report Date', 'Market Center','Arrivals(in tonnes)']]
    dict1[str(2012+i)] = dict1[str(2012+i)].sort_values(['# Commodity', 'State' , 'Report Date', 'Market Center'])
    data = dict1[str(2012+i)].drop('Market Center', axis=1)
    blocks = data.groupby(['# Commodity','State','Report Date'])
    result = blocks.sum().reset_index()
    dict1[str(2012+i)] = dict1[str(2012+i)].merge(result, left_on=['# Commodity', 'State' , 'Report Date'], right_on=['# Commodity', 'State' , 'Report Date'],
                           how='left')
    dict1[str(2012+i)]['Market Share'] = dict1[str(2012+i)]['Arrivals(in tonnes)_x']/dict1[str(2012+i)]['Arrivals(in tonnes)_y']
    dict1[str(2012+i)] = dict1[str(2012+i)].rename(columns={"Arrivals(in tonnes)_y":"Total Market Arrival"})
```

```python
data = dict1['2012']
for i in range(8):
    data = data.append(dict1[str(2013+i)])
    
data
```

pdata - dataframe grouped by Commodity name and State only -- giving a statewise annual report 

```python
g = dict1[str(2012)].drop(['Market Center','Market Share','Report Date','Total Market Arrival'], axis=1).groupby(['# Commodity','State'])
df = g.sum().reset_index()
df['Year'] = [str(2012) for n in range(df.shape[0])]

pdata = df

for i in range(8):
    g = dict1[str(2013+i)].drop(['Market Center','Market Share','Report Date','Total Market Arrival'], axis=1).groupby(['# Commodity','State'])
    df = g.sum().reset_index()
    df['Year'] = [str(2013+i) for n in range(df.shape[0])]
    pdata = pdata.append(df)
    
pdata = pdata.reset_index()
pdata
```

```python
pdata.to_csv('statewise_annual.csv',index=False)
```

adata - dataframe grouped by Commodity name only -- giving a nationwisewise annual report 

```python
g = dict1[str(2012)].drop(['Market Center','Market Share','Report Date','Total Market Arrival','State'], axis=1).groupby(['# Commodity'])
df = g.sum().reset_index()
df['Year'] = [str(2012) for n in range(df.shape[0])]

adata = df

for i in range(8):
    g = dict1[str(2013+i)].drop(['Market Center','Market Share','Report Date','Total Market Arrival','State'], axis=1).groupby(['# Commodity'])
    df = g.sum().reset_index()
    df['Year'] = [str(2013+i) for n in range(df.shape[0])]
    adata = adata.append(df)
    
adata = adata.reset_index()
adata['Arrivals(in tonnes)_x'] = adata['Arrivals(in tonnes)_x'].apply(lambda x: '%.17f' % x).values.tolist()
```
