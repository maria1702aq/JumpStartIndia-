```python
import pandas as pd
from pyjarowinkler import distance
```

```python
print(distance.get_jaro_distance("hello", "haloa", winkler=True, scaling=0.1))

print(distance.get_jaro_distance("hello", "haloa", winkler=False, scaling=0.1))
```

```python
df = pd.read_csv(r'crop_years.csv')
l = list(df.crops)
l.insert(0,0)
```

```python
m=len(l)
print(m)
nlist = [n+1 for n in range(m-1)]
matrix = [l]
```

```python

for i in nlist:
    new_row = [l[i]]
    for num in nlist:
        new_row.append(distance.get_jaro_distance(l[i],l[num], winkler=True, scaling=0.1))
        
    matrix.append(new_row)
```

```python
print(matrix)
```

```python
import numpy as np
np.savetxt("matrix.txt", matrix,fmt='%s')
```
